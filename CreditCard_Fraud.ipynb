```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#To download the dataset, from kaggle, you can use the following command:
# kaggle datasets download mlg-ulb/creditcardfraud

# Import the dataset from the 'creditcard.csv' file into a DataFrame
df = pd.read_csv('creditcard.csv')

# Display the first 5 rows of the dataset to get an initial view of the data
df.head()
```

```python
# Check for missing (null) values in each column of the dataset
df.isnull().sum()
```

```python
# Generate descriptive statistics of the dataset including count, mean, std deviation, min, and max values for each numerical column
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler

# Normalize the 'Amount' feature to have mean 0 and standard deviation 1
df['Amount'] = StandardScaler().fit_transform(df[['Amount']])

# Drop the 'Time' column as it is not likely to contribute meaningfully to the classification
df.drop(['Time'], axis=1, inplace=True)

```

```python
# Count the number of transactions in each class (0 = genuine, 1 = fraudulent)
# This helps identify if the dataset is imbalanced, which is important for model training
df['Class'].value_counts()
```

```python
# Import SMOTE (Synthetic Minority Over-sampling Technique) for handling class imbalance
from imblearn.over_sampling import SMOTE

# Separate features (X) and target variable (y)
X = df.drop('Class', axis=1)  # Drop the 'Class' column to get feature set
y = df['Class']               # Target variable (0 = genuine, 1 = fraud)

# Apply SMOTE to generate synthetic samples for the minority class (fraud cases)
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)  # X_res and y_res are the balanced datasets

```

```python
from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize the class distribution after applying SMOTE
sns.countplot(x=y_res)  # Creates a bar plot showing the count of each class in the resampled dataset
plt.title("Balanced Class Distribution After SMOTE")  # Title for the plot
plt.xlabel("Class")  # Label for x-axis: 0 (genuine), 1 (fraud)
plt.ylabel("Count")  # Label for y-axis: number of samples
plt.show()  # Display the plot

```

```python
from sklearn.model_selection import train_test_split

# Split the resampled dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_res,                # Features after SMOTE
    y_res,                # Target after SMOTE
    test_size=0.2,        # 20% of the data will be used for testing
    random_state=42,      # Ensures reproducibility of the split
    stratify=y_res        # Preserves the class distribution in both train and test sets
)

```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the Random Forest Classifier with 100 decision trees
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model on the training data
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Make predictions using the trained model on the test set
y_pred = model.predict(X_test)

# Print the confusion matrix to evaluate the model's performance
# It shows the counts of True Positives, False Positives, True Negatives, and False Negatives
print(confusion_matrix(y_test, y_pred))

# Print a detailed classification report with precision, recall, and F1-score for each class
# The 'digits=4' argument specifies 4 decimal places for the output
print(classification_report(y_test, y_pred, digits=4))
```

```python
# Generate the classification report as a dictionary
report = classification_report(y_test, y_pred, output_dict=True)

# Convert the dictionary into a pandas DataFrame for better visualization
df_report = pd.DataFrame(report).transpose()

# Set the figure size for the heatmap
plt.figure(figsize=(8, 4))

# Plot a heatmap of the classification report (excluding the 'accuracy' row and column)
sns.heatmap(df_report.iloc[:-1, :-1], annot=True, cmap="YlGnBu")

# Set the title for the heatmap
plt.title("Classification Report")

# Display the heatmap
plt.show()
```

```python
# Generate the confusion matrix to evaluate the model's performance
cm = confusion_matrix(y_test, y_pred)

# Set the figure size for the confusion matrix heatmap
plt.figure(figsize=(6, 4))

# Plot the heatmap of the confusion matrix
# 'annot=True' will annotate the cells with the numeric values
# 'fmt="d"' ensures that the annotations are in integer format
# 'cmap="Blues"' gives the heatmap a blue color scheme
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")

# Set the title for the heatmap
plt.title("Confusion Matrix")

# Set the labels for the x and y axes
plt.xlabel("Predicted")
plt.ylabel("Actual")

# Display the confusion matrix heatmap
plt.show()
```

```python
# Predict the probabilities for each class
# `model.predict_proba(X_test)` returns an array with probabilities for both classes (fraud and genuine)
# We extract the probability for class '1' (fraudulent transactions) by selecting the second column (index 1)
y_proba = model.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Calculate the False Positive Rate (FPR) and True Positive Rate (TPR) for various thresholds
fpr, tpr, _ = roc_curve(y_test, y_proba)  # y_proba contains predicted probabilities of the positive class (fraudulent transactions)

# Calculate the Area Under the Curve (AUC) score to evaluate the model's overall performance
auc = roc_auc_score(y_test, y_proba)  # AUC ranges from 0 to 1, with a higher value indicating a better model

# Plot the ROC curve
plt.figure()  # Create a new figure for the plot
plt.plot(fpr, tpr, label=f'AUC = {auc:.4f}')  # Plot the ROC curve with the AUC value in the label
plt.plot([0, 1], [0, 1], 'k--')  # Add a diagonal dashed line indicating random classifier (AUC = 0.5)
plt.xlabel('False Positive Rate')  # Label for x-axis (False Positive Rate)
plt.ylabel('True Positive Rate')  # Label for y-axis (True Positive Rate)
plt.title('ROC Curve')  # Title for the plot
plt.legend(loc="lower right")  # Display the AUC in the legend at the lower-right corner
plt.grid(True)  # Display gridlines for better readability
plt.show()  # Display the plot

```

```python
from sklearn.metrics import precision_recall_curve

# Calculate precision and recall for different threshold values
precision, recall, _ = precision_recall_curve(y_test, y_proba)  # y_proba contains predicted probabilities of the positive class (fraudulent transactions)

# Plot the Precision-Recall curve
plt.figure()  # Create a new figure for the plot
plt.plot(recall, precision, marker='.')  # Plot precision vs recall for different threshold values
plt.xlabel('Recall')  # Label for the x-axis (Recall)
plt.ylabel('Precision')  # Label for the y-axis (Precision)
plt.title('Precision-Recall Curve')  # Title for the plot
plt.grid(True)  # Display gridlines for better readability
plt.show()  # Display the plot

```
